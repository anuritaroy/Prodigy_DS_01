```python
latest_year ='2023'
```

```python
pop_2023 = df[['Country Name', latest_year]].dropna()

# Sort by population
pop_2023_sorted = pop_2023.sort_values(by=latest_year, ascending=False)
```

```python
pop_2023_asc = df[['Country Name', latest_year]].dropna()

# Sort by population
pop_2023_sorted_asc = pop_2023.sort_values(by=latest_year, ascending=True)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=pop_2023_sorted.head(10), x='Country Name', y=latest_year, palette='viridis')

plt.title(f"Top 10 Most Populous Countries in {latest_year}")
plt.ylabel("Population")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram of population distribution across countries
plt.figure(figsize=(10, 5))
sns.histplot(pop_2023_sorted[latest_year], bins=30, kde=True, color='teal')

plt.title(f"Population Distribution Across Countries in {latest_year}")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```

```python
meta_df = pd.read_csv("C:\\Users\\amiro\Desktop\\Internship Prodigy\\API_SP.POP.TOTL_DS2_en_csv_v2_26346\\Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_26346.csv")

# Merge population data with metadata
merged = pd.merge(pop_2023, meta_df[['Country Code', 'Region']], 
                  left_on='Country Code', right_on='Country Code', how='left')

# Group by region and plot
region_group = merged.groupby('Region')[latest_year].sum().reset_index()

# Plot population by region
plt.figure(figsize=(12, 6))
sns.barplot(data=region_group.sort_values(by=latest_year, ascending=False), 
            x='Region', y=latest_year, palette='coolwarm')
plt.title(f"Population by Region in {latest_year}")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
